```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

## simpe bar plot

```python

sns.countplot(data=df, x='Survived')
plt.title('Survival Count on Titanic')
plt.show()
```

### Line chart showing Age

```python
plt.title('Line chart showing the Age')

sns.lineplot(data=df['Age'], label='Age')
```

### Scatter plot showing the relationship between Fare charged and Age

```python
plt.title('Relationship between Fare charged and Age')
sns.scatterplot(x = 'Fare', y ='Age', data = df)
```
